```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'C:/python/data.csv'

# CSV 파일에서 데이터 불러오기
try:
    df = pd.read_csv(csv_file_path, encoding='cp949')
except FileNotFoundError:
    print(f"파일 '{csv_file_path}'을 찾을 수 없습니다.")
except Exception as e:
    print(f"데이터를 불러오는 중 오류가 발생했습니다: {e}")

# '총매출'과 '목표매출' 데이터 비교
sales_data = df.loc[[0, 1], '9월':'2월']

# ','와 '$' 제거 및 숫자로 변환
sales_data = sales_data.replace('[\$,]', '', regex=True).astype(float)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(sales_data.T, marker='o')
plt.title('매출과 매출목표')
plt.xlabel('월')
plt.ylabel('금액')
plt.legend(['총매출', '목표매출'])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'C:/python/data.csv'

# CSV 파일에서 데이터 불러오기
try:
    df = pd.read_csv(csv_file_path, encoding='cp949')
except FileNotFoundError:
    print(f"파일 '{csv_file_path}'을 찾을 수 없습니다.")
except Exception as e:
    print(f"데이터를 불러오는 중 오류가 발생했습니다: {e}")

# '광고비용'과 '소셜네트워크 비용' 데이터 비교
ad_cost_data = df.loc[2, '9월':'2월']
social_network_cost_data = df.loc[3, '9월':'2월']

# '$' 제거 및 숫자로 변환
ad_cost_data = ad_cost_data.replace('[\$,]', '', regex=True).astype(float)
social_network_cost_data = social_network_cost_data.replace('[\$,]', '', regex=True).astype(float)

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(ad_cost_data, marker='o', label='광고비용', color='blue')
plt.plot(social_network_cost_data, marker='o', label='소셜네트워크 비용', color='orange')
plt.title('광고비용과 소셜네트워크 비용 변화')
plt.xlabel('월')
plt.ylabel('비용 ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

csv_file_path = 'C:/python/data.csv'

# CSV 파일에서 데이터 불러오기 
try:
    df = pd.read_csv(csv_file_path, encoding='cp949')
except FileNotFoundError:
    print(f"파일 '{csv_file_path}'을 찾을 수 없습니다.")
except Exception as e:
    print(f"데이터를 불러오는 중 오류가 발생했습니다: {e}")

# '총매출'과 '목표매출' 데이터
sales_data = df.loc[[0, 1], '9월':'2월']

# ','와 '$' 제거 및 숫자로 변환
sales_data = sales_data.replace('[\$,]', '', regex=True).astype(float)

# 경쟁사 예상 매출 (가정)
competitor_expected_sales = sales_data.iloc[0, :] * 1.5  # 예상 매출을 현재 매출의 1.5배로 가정

# 자사 매출 목표
own_sales_target = sales_data.iloc[1, :]

# 자사 매출
own_sales = sales_data.iloc[0, :]

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 천 단위 구분 기호 표시 함수 정의
def price_formatter(x, pos):
    return "${:,.0f}".format(x)

# y축에 포맷 적용
formatter = FuncFormatter(price_formatter)

plt.plot(own_sales, marker='o', label='자사 매출', color='blue')
plt.plot(own_sales_target, marker='o', label='자사 매출 목표', color='green', linestyle='--')
plt.plot(competitor_expected_sales, marker='o', label='경쟁사 예상 매출', color='red', linestyle='--')
plt.title('자사 매출과 경쟁사 예상 매출 비교')
plt.xlabel('월')
plt.ylabel('매출액')
plt.legend()
plt.gca().get_yaxis().set_major_formatter(formatter)  # y축 포맷 적용
plt.grid(True)
plt.show()
```
